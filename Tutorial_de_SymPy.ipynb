## <font color=red> Matrices con la libreria SymPy

# <font color = bluer> libreria SymPy


#SymPy es una biblioteca de Python para matemática simbólica. Permite trabajar con matrices de manera muy similar a como lo harías en papel, incluyendo operaciones como multiplicación, potencias, cálculo de autovalores y autovectores, etc. En este tutorial veremos los aspectos fundamentales para manejar matrices en SymPy.

##<font color = Purple> 1.- Importar la librería

Primero debemos importar SymPy:
##



```python
import sympy as sp
```

##<font color = Purple>2.-Crear y representar matrices

En SymPy, las matrices se crean con sp.Matrix().
Se escriben como una lista de listas, donde cada lista interna representa un renglón.

```python
# Definimos una matriz 2x2
A = sp.Matrix([ [1, 2],
                [3, 4] ])

A
```

También podemos usar elementos simbólicos:

```python
x, y = sp.symbols('x y')

B = sp.Matrix([ [x, 1],
                [y, 2] ])

B
```

##<font color = Purple>3.-Obtener una entrada específica

En Python, los índices comienzan en 0.

Obtener un elemento específico

Para obtener el elemento en la fila i y columna j:

```python
elemento = A[0, 1]   # Fila 0, columna 1
print("Elemento en fila 0, columna 1:", elemento)
```

Obtener un renglón completo

```python
#row(i)devuelve el renglón (fila) número i de la matriz.
# Importante: los índices empiezan en 0.
# Es decir:
# row(0) -> primera fila
# row(1) -> segunda fila
renglon = A.row(0)
reglon2 = A.row(1)

print("Primer renglón:", renglon)
print("Segundo renglón:", reglon2)
```

Obtener una columna completa

```python
#A.col(j) devuelve la columna número j de la matriz.
# También empieza en índice 0.
# Es decir:
# col(0) -> primera columna
# col(1) -> segunda columna
columna = A.col(0)
columna2 = A.col(1)

print("Primera columna:", columna)
print("Segunda columna:", columna2)
```

##<font color = Purple>4.-Multiplicación de matrices

La multiplicación se hace con el operador $*$.

```python
producto = A * B

print("Producto A * B:")

producto
```

# **NOTA**:Para poder multiplicar matrices, el número de columnas de la primera debe ser igual al número de renglones de la segunda.

##<font color = Purple>5.-Elevar una matriz a una potencia

Para elevar una matriz a una potencia usamos $**$.

```python
A_cuadrada = A**2

print("A al cuadrado:")
A_cuadrada
```

También se puede elevar a cualquier potencia entera positiva:

```python
A_cubo = A**3

print("A al cubo:")
A_cubo
```

#**NOTA** Matemáticamente, elevar una matriz significa multiplicarla por sí misma varias veces.

##<font color = Purple>6.-Multiplicar vectores por matrices

En SymPy, los vectores se representan como matrices columna.

##<font color = red>Vector columna

```python
v = sp.Matrix([
    [1],
    [2]
])

resultado = A * v

print("Resultado de A * v:")
resultado
```

##<font color = red>Vector fila

```python
v_fila = sp.Matrix([[1, 2]])

resultado_fila = v_fila * A

print("Resultado de v_fila * A:")

resultado_fila
```

##<font color = Purple>7.-Valores propios (eigenvalores)

Los valores propios se calculan con:

```python
valores_propios = A.eigenvals()

print("Valores propios de A:")
print(valores_propios)
```

##<font color = Purple>8.-Vectores propios (eigenvectores)

Para obtener los vectores propios:

```python
vectores_propios = A.eigenvects()

print("Vectores propios de A:")
for valor, multiplicidad, vectores in vectores_propios:
    print("\nValor propio:", valor)
    print("Multiplicidad:", multiplicidad)
    print("Vectores propios:", vectores)
```
