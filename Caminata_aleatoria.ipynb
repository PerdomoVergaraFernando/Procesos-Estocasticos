Implemente una camita aleatoria $1D$, con probabilidad $p$ de moverse a la derecha y probabilidad q de moverse a la izquierda, tal que $p + q = 1$.

Importamos la libreria random


```python
import random as r
import matplotlib.pyplot as plt
import numpy as np
```

```python
def caminata_aleatoria(p, N_pasos=100, n_trayectorias=10):

    # Generar matriz de incrementos: +1 con prob p, -1 con prob 1-p
    incrementos = np.where(np.random.rand(n_trayectorias, N_pasos) < p, 1, -1)

    # Acumular incrementos para obtener la posición en cada paso
    posiciones = np.zeros((n_trayectorias, N_pasos + 1))
    posiciones[:, 1:] = np.cumsum(incrementos, axis=1)

    return posiciones
```

```python
# Parámetros
p_fijo = .5
N = 100

# Simular
trayectorias = caminata_aleatoria(p_fijo, N)

# Graficar
plt.figure(figsize=(10, 6))
plt.plot(trayectorias[0], lw=1, alpha=0.7) # Traza solo la primera trayectoria
plt.axhline(0, color='red', linestyle='--', alpha=0.8) # Add horizontal line at y=0
plt.xlabel('Paso')
plt.ylabel('Posición')
plt.title(f'Caminata aleatoria 1D con p = {p_fijo} (Una trayectoria)')
plt.grid(alpha=0.2)
plt.show()
```
