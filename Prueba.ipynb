Implemente una camita aleatoria $1D$, con probabilidad $p$ de moverse a la derecha y probabilidad q de moverse a la izquierda, tal que $p + q = 1$



```python
import random
import matplotlib.pyplot as plt

def caminata_aleatoria(n_pasos=100, p=0.5):

    if p < 0 or p > 1:
        raise ValueError("p debe estar entre 0 y 1")

    # Inicia en 0
    posiciones = [0]

    for i in range(n_pasos):
        r = random.random()  # número uniforme en [0,1)

        if r < p:
            nuevo = posiciones[-1] + 1
        else:
            nuevo = posiciones[-1] - 1

        posiciones.append(nuevo)

    return posiciones

# Parámetros
n = 200
p = 0.5   # Puedes cambiar este valor

trayectoria = caminata_aleatoria(n, p)

# Gráfica (solo una línea)
plt.figure()
plt.plot(trayectoria)
plt.title(f"Caminata Aleatoria 1D (p = {p})")
plt.xlabel("Paso")
plt.ylabel("Posición")
plt.axhline(0)
plt.show()
```
