#<font color = Purple> 2.- Crear un Grafo vacio.

## Tipos de Graficos

Graph(): Funcion para crea un grafo no dirigido.

DiGraph(): Funcion para crea un grafo dirigido.

MultiGraph(): Funcion para crea un grafo no dirigido con múltiples aristas entre nodos.

```python
G=nx.Graph() #Se lo asigmanos a una variable para poder trabajar mejor.
```

#<font color = Purple> 3.- Agregar Nodos.

add_node(nodo): Funcion que agrega un único nodo.

add_nodes_from(lista): Funcion que permite añadir varios nodos de golpe.

Los nodos pueden llevar atributos, como color, peso, nombre, etc. (se guardan como diccionarios asociados al nodo).

vea el siguiente ejemplo:


```python
# Agregar un nodo individual
G.add_node(1)

# Agregar varios nodos a la vez
G.add_nodes_from([2, 3])

# Agregar nodos con atributos (ejemplo: color)
G.add_nodes_from([(4, {"color": "rojo"}), (5, {"color": "verde"})])

nx.draw_networkx(G)
```

#<font color = Purple> 4.- Agregar Aristas.

Una arista conecta dos nodos.

add_edge(a, b): conecta a con b.

add_edges_from(lista): permite meter muchas aristas a la vez.

En este caso, ahora tenemos un grafo con 5 nodos y varias conexiones.

vea el siguiente ejemplo:

```python
# Agregar una arista entre 1 y 2
G.add_edge(1, 2)

# Agregar usando tupla
e = (2, 3)
G.add_edge(*e)

# Agregar múltiples aristas
G.add_edges_from([(1, 3), (3, 4), (4, 5)])
nx.draw_networkx(G)
```

#<font color = Purple>  5.- Explorar el Grafo.

G.nodes: Funcion que devuelve los nodos.

G.edges: Funcion que devuelve las conexiones.

G.adj[nodo]: Funcion que muestra los vecinos de un nodo.

G.degree[nodo]: Funcion que muestra cuántas aristas tiene ese nodo.

nx.number_of_nodes(): Está función me devuelve la cantidad de nodos.

nx.number_of_edges() # Está función me devuelve la cantidad de Aristas (Arcos, Ramas o Ligaduras).

nx.shortest_path(G, source, target, weight=None): Funcion que me da el camino mas corto del Grafo.

      G: el grafo.

      source: nodo de inicio.

      target: nodo de destino.

      weight: si quieres usar pesos (ej: distancia, tiempo, costo).

      weight=None → usa la cantidad de aristas (camino más corto en saltos).

      weight="nombre_atributo" → usa el atributo que hayas definido en las aristas (ej: "distancia").

vea el siguiente ejemplo:

```python
# Listar nodos y aristas
print("Nodos:", G.nodes)
print("Aristas:", G.edges)

# Ver nodos adyacentes a 1
print("Vecinos de 1:", list(G.adj[1]))

# Ver el grado de un nodo (cuántas conexiones tiene)
print("Grado de 1:", G.degree[1])

print("Cantidad de nodos", nx.number_of_nodes(G)) #Está función me devuelve la cantidad de nodos

print("Cantidad de Aristas",nx.number_of_edges(G)) # Está función me devuelve la cantidad de Aristas (Arcos, Ramas o Ligaduras)

print("El camino mas corto", nx.shortest_path(G, 1, 5, weight=None))
```

#<font color = Purple>  6.- Eliminar Nodos y Aristas

Puedes borrar nodos y todas sus aristas.

También puedes borrar solo ciertas conexiones sin eliminar el nodo

remove_node(nodo): Funcion que elimina un nodo.

G.remove_nodes_from([a, b]): Funcion que elimina varios nodos.

G.remove_edge(a, b): Funcion que elimina una arista.

vea el siguiente ejemplo:

```python
# Eliminar un nodo
G.remove_node(5)

# Eliminar varios nodos
# Check if nodes 2 and 3 still exist before attempting to remove them
nodes_to_remove = [node for node in [2, 3] if node in G]
G.remove_nodes_from(nodes_to_remove)

# Eliminar una arista
# Check if the edge (1, 3) still exists before attempting to remove it
if G.has_edge(1, 3):
    G.remove_edge(1, 3)

nx.draw_networkx(G)
```

#<font color = Purple>  6.- Constrir Grafos de otra forma.

En lugar de agregar nodos y aristas paso a paso, también podemos crear grafos directamente:

Graph([(a, b), (b, c), ...]) Funcion que construye un grafo directamente con las conexiones.

Graph(diccionario): Funcion que usa un diccionario de adyacencia, donde cada clave es un nodo y su valor es la lista de vecinos.

vea los siguientes ejemplos:

```python
# Grafo a partir de una lista de aristas
H = nx.Graph([(0, 1), (1, 2), (2, 3)])

nx.draw_networkx(H)
```

```python
# Grafo desde diccionario de adyacencia
adj_dict = {0: [1, 2], 1: [0, 2], 2: [0, 1]}
J = nx.Graph(adj_dict)
nx.draw_networkx(J) #Funcion que dibuja el grafo.
```

#<font color = Purple>  7.- Visualizacion basica.

NetworkX tiene funciones sencillas para dibujar.

vea el siegiente ejemplo.

# <font color= red> Ejemplos que hemos hecho en clases

```python
import networkx as nx
import matplotlib.pyplot as plt

# Crear grafo dirigido
F = nx.DiGraph()

# Estados
F.add_nodes_from(["0", "1"])

# Transiciones con probabilidad
F.add_edge("0", "0", prob=0.8)
F.add_edge("0", "1", prob=0.2)

F.add_edge("1", "0", prob=0.6)
F.add_edge("1", "1", prob=0.4)


# Posiciones de los nodos
pos = nx.spring_layout(F, seed=8)

plt.figure(figsize=(6,4))

# Dibujar nodos
nx.draw_networkx_nodes(F, pos, node_size=1400)

# Etiquetas de nodos
nx.draw_networkx_labels(F, pos, font_size=12)

# Separar loops y aristas normales
loops = [(u,v) for u,v in F.edges() if u==v]
edges = [(u,v) for u,v in F.edges() if u!=v]

# Dibujar aristas normales
nx.draw_networkx_edges(
    F,
    pos,
    edgelist=edges,
    arrowstyle='-|>',
    arrowsize=18,
    width=2,
    connectionstyle="arc3,rad=0.15"
)

# Dibujar loops (más curvos para que se vean)
nx.draw_networkx_edges(
    F,
    pos,
    edgelist=loops,
    arrowstyle='-|>',
    arrowsize=18,
    width=2,
    connectionstyle="arc3,rad=0.35"
)

# Etiquetas de probabilidades
edge_labels = {(u, v): f"p={d['prob']}" for u, v, d in F.edges(data=True)}

nx.draw_networkx_edge_labels(
    F,
    pos,
    edge_labels=edge_labels,
    font_size=11
)

plt.axis("off")
plt.title("Diagrama de transición")
plt.show()
```

```python
import matplotlib.pyplot as plt

F = nx.DiGraph()

# Estados
F.add_nodes_from(["0", "1", "2"])

# Transiciones con probabilidad
F.add_edge("0", "0", prob=0.5)
F.add_edge("0", "1", prob=0.3)
F.add_edge("0", "2", prob=0.2)

F.add_edge("1", "1", prob=0.2)
F.add_edge("1", "0", prob=0.5)
F.add_edge("1", "2", prob=0.3)

F.add_edge("2", "1", prob=0.5)
F.add_edge("2", "0", prob=0.4)
F.add_edge("2", "2", prob=0.1)

# Dibujar
pos = nx.spring_layout(F, seed=8)

plt.figure(figsize=(6, 4))
nx.draw_networkx_nodes(F, pos, node_size=1400)
nx.draw_networkx_labels(F, pos, font_size=12)
nx.draw_networkx_edges(F, pos, arrowstyle='-|>', arrowsize=18, width=2)

# Etiquetas en aristas
edge_labels = {(u, v): f"p={d['prob']}" for u, v, d in F.edges(data=True)} # Changed G to F here
nx.draw_networkx_edge_labels(F, pos, edge_labels=edge_labels, font_size=11)

plt.axis("off")
plt.title("Diagrama de transición")
plt.show()
```
